```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.fill_between(
    x, stats.norm.pdf(x, mu, sigma),
    where=(x < mu - sigma),
    color="gray",
    alpha=0.2
)
plt.axis("off")
plt.show()
```
